```python
import numpy as np
from scipy.signal import correlate2d
from PIL import Image
import random
```

```python
#Importing the image and making a 2d numpy array out of it with the 
pic = Image.open('example-size.png').convert('L')
np_pic = np.array(pic)
#This line shows the image
Image.fromarray(np.uint8(np_pic * 255), 'L')
```

```python
#This is an individual patch average function
def patch_avg(x,y,a):
    r = (1/8)
    kernel = np.array([[r, r, r],
                       [r, 0, r],
                       [r, r, r]])
    c = correlate2d(a, kernel, mode='same')
    return c[x][y]
```

```python
#This function replaces any pixel with value 0 with it's patch average
def fill_holes(a):
    for i in range(a.shape[0]):
        for j in range(a.shape[1]):
            if a[i][j] == 0:
                a[i][j] = patch_avg(i,j,a)
```

```python
#This portion fills the image with holes randomly
for i in range(np_pic.shape[0]):
    for j in range(np_pic.shape[1]):
        num = random.randint(1,51)
        if num == 1:
            np_pic[i][j] = 0
```

```python
#Shows the holey image
Image.fromarray(np.uint8(np_pic * 255)).show()
```

```python
#Fills the holes and displays the result
fill_holes(np_pic)
Image.fromarray(np.uint8(np_pic * 255)).show()
```
